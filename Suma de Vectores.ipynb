1. **Configuración de los límites del gráfico**: Se establecen los límites de los ejes X y Y para que vayan desde -8 hasta 8.
2. **Habilitar la cuadrícula**: Se activa una cuadrícula para mejorar la visibilidad de las líneas y componentes.
3. **Título**: El gráfico se titula "Suma de vectores".
4. **Vector A**: Se grafica el vector A a partir del origen (0,0)
5. 4. **Vector B y C**: Se grafican los vectores con origen en el final del vector de la letra anterior
6. **Vector A+B+C** Se grafica desde el origen la suma de vectores
7. **Etiquetas**: Se añaden etiquetas a los  vectores
8. **Mostrar el gráfico**: Finalmente, el gráfico se renderiza utilizando `plt.show()`.

```python
# Definir límites de los ejes del plano
x1 = -8
x2 = 8
y1 = -8
y2 = 8
plt.axis([x1, x2, y1, y2])

# Configuración de los ejes
plt.axis('on')
plt.grid(True)
plt.title("Suma de vectores")

# Definir Cada vector por medio de origen y desplazamiento
origin_Ax = 0
origin_Ay = 0
despAx = 2
despAy = 4

origin_Bx = origin_Ax + despAx
origin_By = origin_Ay + despAy
despBx = -6
despBy = -2

origin_Cx = origin_Bx + despBx
origin_Cy = origin_By + despBy
despCx = 2
despCy = -7

origin_SUMx = 0
origin_SUMy = 0
despSUMx = despAx + despBx + despCx
despSUMy = despAy + despBy + despCy

# Graficar el cada vector por medio de las variables de origen y desplazamiento
plt.arrow(origin_Ax, origin_Ay, despAx, despAy, head_length=0.2, head_width=0.3, color="red")
plt.arrow(origin_Bx, origin_By, despBx, despBy, head_length=0.2, head_width=0.3, color="blue")
plt.arrow(origin_Cx, origin_Cy, despCx, despCy, head_length=0.2, head_width=0.3, color="green")
plt.arrow(origin_SUMx, origin_SUMy, despSUMx, despSUMy, head_length=0.2, head_width=0.3, color="black")


# Añadir etiquetas a las componentes
plt.text(2, 2, r'$\vec{A}<2,4>$', size=15, color='red')
plt.text(-3.5, 4, r'$\vec{B}<-6,-2>$', size=15, color='blue')
plt.text(-7.9, 0, r'$\vec{C}<2,-7>$', size=15, color='green')
plt.text(-1, -4, r'$\vec{A+B+C}<-2,-5>$', size=15, color='black')

# Mostrar la gráfica
plt.show()
```
