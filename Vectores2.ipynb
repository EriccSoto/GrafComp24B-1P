# Vector A (3,4)

El código utiliza la biblioteca `matplotlib` para graficar un vector A = (3, 4) con origen en el punto (2, 3) en un gráfico bidimensional:

1. **Configuración de los límites del gráfico**: Se establecen los límites de los ejes X y Y para que vayan desde -1 hasta 10.
2. **Habilitar la cuadrícula**: Se activa una cuadrícula para mejorar la visibilidad de las líneas y componentes.
3. **Título**: El gráfico se titula "Vector A".
4. **Vector A**: Se grafica el vector A a partir del punto (2, 3) con un desplazamiento en X = 3 y Y = 4.
5. **Componentes del vector**: Se dibujan líneas punteadas que representan las proyecciones del vector en los ejes X (en rojo) y Y (en azul).
6. **Etiquetas**: Se añaden etiquetas al vector A y a sus componentes en X y Y.
7. **Mostrar el gráfico**: Finalmente, el gráfico se renderiza utilizando `plt.show()`.

```python
import matplotlib.pyplot as plt

# Definir límites del plano
x1 = -1
x2 = 10
y1 = -1
y2 = 10
plt.axis([x1, x2, y1, y2])

# Configuración de los ejes
plt.axis('on')
plt.grid(True)
plt.title("Vector A") #titulo

# Definir el vector A
dx = 3
dy = 4

# Graficar el vector A
plt.arrow(2, 3, dx, dy, head_length=0.2, head_width=0.2, color="black")

# Graficar el componente en el eje x (línea punteada en rojo)
plt.plot([2, 5], [3, 3], linestyle='--', color='red')

# Graficar el componente en el eje y (línea punteada en azul)
plt.plot([5, 5], [3, 7], linestyle='--', color='blue')

# Añadir etiquetas a las componentes
plt.text(3.7, 6, r'$\vec{A}$', size=15, color='black')
plt.text(5.5, 4.8, r'$\vec{Ay}$', size=15, color='blue')
plt.text(3.5, 2, r'$\vec{Ax}$', size=15, color='red')

# Mostrar la gráfica
plt.show()
```

# Vector B (-5,2)

El código grafica el vector B = (-5, 2) en un plano cartesiano con origen en el punto (2, -3), y representa visualmente sus componentes en los ejes X y Y con líneas punteadas:

1. **Configuración de los ejes del plano**: Se definen los límites del gráfico, configurando los ejes X desde -5 hasta 4 y Y desde -4 hasta 3.
2. **Habilitar la cuadrícula**: Se activa la cuadrícula para facilitar la visualización de las componentes.
3. **Título**: El gráfico tiene el título "Vector B".
4. **Vector B**: Se grafica el vector B con un desplazamiento de -5 en el eje X y 2 en el eje Y a partir del punto (2, -3).
5. **Componentes del vector**: Se dibujan líneas punteadas que representan las proyecciones del vector en los ejes X (en rojo) y Y (en azul).
6. **Etiquetas**: Se añaden etiquetas al vector B y a sus componentes en X y Y con colores correspondientes.
7. **Mostrar la gráfica**: Finalmente, se renderiza el gráfico con la función `plt.show()`.

```python
# Se definen los ejes del plano
x1 = -5
x2 = 4
y1 = -4
y2 = 3

# Configuración de los ejes
plt.axis([x1, x2, y1, y2])

plt.axis('on')
plt.grid(True)
plt.title("Vector B")

# Definir el vector B
dx = -5
dy = 2

# Graficar el vector B
plt.arrow(2, -3, dx, dy, head_length=0.2, head_width=0.2, color="Black")

# Dibuja el componente en el eje y en línea punteada
plt.plot([-3, -3], [-3, -1], linestyle='--', color='blue')

# Dibuja el componente en el eje x en línea punteada
plt.plot([2, -3], [-3, -3], linestyle='--', color='red')

# Añadir etiquetas a las componentes
plt.text(-1.8, -1.5, r'$\vec{B}$', size=15, color='black')
plt.text(-1, -3.5, r'$\vec{Bx}$', size=15, color='red')
plt.text(-3.5, -2, r'$\vec{By}$', size=15, color='blue')

# Mostrar la gráfica
plt.show()
```

# Vector C (2,-4)

El código grafica el vector **C = (2, -4)** en un plano cartesiano con origen en el punto **(2, -3)**, y representa visualmente sus componentes en los ejes **X** y **Y** mediante líneas punteadas:

1. **Configuración de los ejes del plano**: Se establecen los límites del gráfico, con el eje **X** desde 0 hasta 5 y el eje **Y** desde -8 hasta 1, para acomodar adecuadamente el vector y sus componentes.
2. **Habilitar la cuadrícula**: Se activa una cuadrícula que mejora la visibilidad de las líneas y componentes del vector.
3. **Título**: El gráfico lleva el título **"Vector C"**.
4. **Vector C**: El vector **C** se dibuja desde el punto **(2, -3)**, con un desplazamiento de 2 en el eje **X** y -4 en el eje **Y**.
5. **Componentes del vector**: Se añaden líneas punteadas que representan las proyecciones del vector en el eje **X** (en rojo) y el eje **Y** (en azul).
6. **Etiquetas**: Se incluyen etiquetas para identificar el vector **C** y sus componentes **Cx** (eje X) y **Cy** (eje Y).
7. **Mostrar la gráfica**: Finalmente, el gráfico se renderiza utilizando la función `plt.show()`.

```python
import matplotlib.pyplot as plt

# Definir límites de los ejes
x1 = 0
x2 = 5
y1 = -8
y2 = 1
plt.axis([x1, x2, y1, y2])

# Configuración de los ejes
plt.axis('on')
plt.grid(True)
plt.title("Vector C")

# Definir el vector C por medio de origen y desplazamiento
origin_x = 2
origin_y = -3
dx = 2
dy = -4

# Graficar el vector C por medio de las variables de origen y desplazamiento
plt.arrow(origin_x, origin_y, dx, dy, head_length=0.2, head_width=0.1, color="black")

# Graficar el componente en el eje x (línea punteada en rojo)
plt.plot([origin_x, origin_x + dx], [origin_y + dy, origin_y + dy], linestyle='--', color='red')

# Graficar el componente en el eje y (línea punteada en azul)
plt.plot([origin_x, origin_x ], [origin_y, origin_y + dy], linestyle='--', color='blue')

# Añadir etiquetas a las componentes
plt.text(3.2, -5, r'$\vec{C}$', size=15, color='black')
plt.text(3, -7.8, r'$\vec{Cx}$', size=15, color='red')
plt.text(1.5, -5, r'$\vec{Cy}$', size=15, color='blue')

# Mostrar la gráfica
plt.show()
```

# Vector D (-7,-3)

El código grafica el vector **D = (-7, -3)** en un plano cartesiano con origen en el punto **(-2, -3)**, y representa visualmente sus componentes en los ejes **X** e **Y** mediante líneas punteadas:

1. **Configuración de los ejes del plano**: Se establecen los límites del gráfico, con el eje **X** desde -10 hasta 2 y el eje **Y** desde -8 hasta 1, para acomodar adecuadamente el vector y sus componentes.
2. **Habilitar la cuadrícula**: Se activa una cuadrícula para mejorar la visibilidad de las líneas y componentes del vector.
3. **Título**: El gráfico lleva el título **"Vector D"**.
4. **Vector D**: El vector **D** se dibuja desde el punto **(-2, -3)**, con un desplazamiento de **-7** en el eje **X** y **-3** en el eje **Y**.
5. **Componentes del vector**: Se añaden líneas punteadas que representan las proyecciones del vector en el eje **X** (en rojo) y el eje **Y** (en azul).
6. **Etiquetas**: Se incluyen etiquetas para identificar el vector **D** y sus componentes **Dx** (eje X) y **Dy** (eje Y).
7. **Mostrar la gráfica**: Finalmente, el gráfico se renderiza utilizando la función `plt.show()`

```python
import matplotlib.pyplot as plt

# Definir límites de los ejes
x1 = -10
x2 = 2
y1 = -8
y2 = 1
plt.axis([x1, x2, y1, y2])

# Configuración de los ejes
plt.axis('on')
plt.grid(True)
plt.title("Vector D")

# Definir el vector D
origin_x = -2
origin_y = -3
dx = -7
dy = -3

# Graficar el vector D usando variables de origen y desplazamiento
plt.arrow(origin_x, origin_y, dx, dy, head_length=0.2, head_width=0.1, color="black")

# Graficar el componente en el eje x (línea punteada en rojo)
plt.plot([origin_x, origin_x + dx], [origin_y + dy, origin_y + dy], linestyle='--', color='red')

# Graficar el componente en el eje y (línea punteada en azul)
plt.plot([origin_x, origin_x ], [origin_y, origin_y + dy], linestyle='--', color='blue')

# Añadir etiquetas a las componentes
plt.text(-6, -4, r'$\vec{D}$', size=15, color='black')
plt.text(-6, -7, r'$\vec{Dx}$', size=15, color='red')
plt.text(-1.72, -5, r'$\vec{Dy}$', size=15, color='blue')

# Mostrar la gráfica
plt.show()
```
