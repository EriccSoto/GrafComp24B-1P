# Salida de 0 a 40 en X y Y

```python
x1 = 0
x2 = 40
y1 = 0
y2 = 40
#Editamos los ejes
plt.axis([x1,x2,y1,y2])
```

#### 1. Definición de los límites de los ejes:
Se definen las variables para controlar los límites de los ejes `x` e `y`:

- `x1 = 0`: Límite inferior del eje `x`.
- `x2 = 40`: Límite superior del eje `x`.
- `y1 = 0`: Límite inferior del eje `y`.
- `y2 = 40`: Límite superior del eje `y`.

Posteriormente, los límites se aplican utilizando la función `plt.axis([x1, x2, y1, y2])


#### 2. Activación de los ejes:
Se activan los ejes del gráfico con `plt.axis('on')`, lo que asegura que los ejes `x` e `y` sean visibles

#### 3. Agregado de la cuadrícula:
Para hacer que la cuadrícula sea visible en el gráfico, se utiliza el comando `plt.grid(True)`, lo que habilita una cuadrícula que ayuda a visualizar mejor la ubicación de los datos en los ejes

#### 4. Añadir un título:
Finalmente, se añade un título descriptivo al gráfico con `plt.title("Ejemplo")`

```python
x1 = 0
x2 = 40
y1 = 0
y2 = 40
#Editamos los ejes
plt.axis([x1,x2,y1,y2])

#Definimos los ejes
plt.axis('on')
plt.grid(True)
plt.title("Ejemplo")
```

# Se agregan Puntos

#### 4. Graficar puntos en el gráfico:
Para graficar puntos en una distribución uniforme, se define un intervalo para los puntos en los ejes `x` e `y` utilizando los valores `dx` y `dy`, que determinan el espacio entre los puntos. En este caso, se ha definido un intervalo de `5` unidades para ambos ejes

Luego, se utiliza un bucle doble `for` para iterar sobre los valores del rango `x1` a `x2` en el eje `x` y de `y1` a `y2` en el eje `y`. En cada iteración, se utiliza `plt.scatter` para dibujar un punto rojo (`color='red'`) con un tamaño (`s=15`) en las coordenadas `(x, y)`

```python
x1 = 0
x2 = 40
y1 = 0
y2 = 40
#Editamos los ejes
plt.axis([x1,x2,y1,y2])

#Definimos los ejes
plt.axis('on')
plt.grid(True)
plt.title("Ejemplo")
dx=5
dy=5
#Graficar puntos
for x in np.arange(x1,x2,dx):
    for y in np.arange(y1,y2,dy):
        #plt.scatter(x_array, 5_tamaño, color, etc)
        plt.scatter(x, y, s = 15, color ='red')
```

En esta versión, se realizaron los siguientes cambios:

1. **Intervalos de los puntos:**
   - **dx = 3**: El intervalo entre los puntos en el eje `x` ahora es de 3 unidades, lo que significa que los puntos estarán más cercanos en este eje.
   - **dy = 7**: El intervalo entre los puntos en el eje `y` es de 7 unidades, lo que hace que los puntos estén más espaciados en este eje en comparación con el eje `x`.

2. **Color de los puntos:**
   - Los puntos ahora se grafican en color **azul** (`color='blue'`) en lugar de rojo, lo que cambia la apariencia visual del gráfico.

```python
x1 = 0
x2 = 40
y1 = 0
y2 = 40
#Editamos los ejes
plt.axis([x1,x2,y1,y2])

#Definimos los ejes
plt.axis('on')
plt.grid(True)
plt.title("Ejemplo")
dx=3
dy=7
#Graficar puntos
for x in np.arange(x1,x2,dx):
    for y in np.arange(y1,y2,dy):
        #plt.scatter(x_array, 5_tamaño, color, etc)
        plt.scatter(x, y, s = 15, color ='blue')
```

# Se Agrega Vector (Se retoma el plano de 0-80)


Este vector creado con `plt.arrow` tiene las siguientes características:

- **Inicio del vector**: El vector comienza en las coordenadas (0, 75).
- **Desplazamiento**: El vector se desplaza 10 unidades en el eje **x** y 0 unidades en el eje **y**. Esto significa que es un vector horizontal hacia la derecha.
- **Longitud de la cabeza del vector** (`head_length=4`): La punta del vector (la flecha) tiene una longitud de 4 unidades.
- **Ancho de la cabeza del vector** (`head_width=4`): El ancho de la punta del vector es de 4 unidades.
- **Color** (`color="k"`): El color del vector es negro, ya que `"k"` es la abreviatura de "black" en `matplotlib`.

En resumen, es un vector horizontal que parte de (0, 75), se extiende 10 unidades hacia la derecha y tiene una cabeza de flecha de tamaño 4x4 en color negro.

```python
x1 = 0
x2 = 80
y1 = 0
y2 = 80
#Editamos los ejes
plt.axis([x1,x2,y1,y2])
#Definimos los ejes
plt.axis('on')
plt.grid(True)
plt.title("Ejemplo")
dx=5
dy=5
#Graficar puntos
for x in np.arange(x1,x2,dx):
    for y in np.arange(y1,y2,dy):
        #plt.scatter(x_array, 5_tamaño, color, etc)
        plt.scatter(x, y, s = 1.5, color ='blue')
plt.arrow(0,75,10,0,  head_length=4, head_width=4,color ="k")
```

Cambio realizado en el vector original 

- **Posición inicial del vector**: La nueva posición inicial del vector es ahora (5, 75) en lugar de (0, 75). Esto significa que el vector empieza 5 unidades más a la derecha en el eje **x**.
- **Ancho de la cabeza del vector** (`head_width=2`): El ancho de la punta del vector ha cambiado de 4 unidades a 2 unidades, haciéndola más delgada.

```python
x1 = 0
x2 = 80
y1 = 0
y2 = 80
#Editamos los ejes
plt.axis([x1,x2,y1,y2])
#Definimos los ejes
plt.axis('on')
plt.grid(True)
plt.title("Ejemplo")
dx=5
dy=5
#Graficar puntos
for x in np.arange(x1,x2,dx):
    for y in np.arange(y1,y2,dy):
        #plt.scatter(x_array, 5_tamaño, color, etc)
        plt.scatter(x, y, s = 1.5, color ='blue')
plt.arrow(5,75,10,0,  head_length=4, head_width=2,color ="k")
```
