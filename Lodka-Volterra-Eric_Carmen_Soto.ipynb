## Parámetros del modelo:

- **Tasa de crecimiento de las presas en ausencia de depredadores (a):** 0.7
- **Tasa de depredación (b):** 0.5
- **Tasa de muerte de depredadores en ausencia de presas (c):** 0.3
- **Tasa a la cual los depredadores aumentan por cada presa consumida (e):** 0.2



```python
%matplotlib inline
import matplotlib.pyplot as plt
from random import *
from numpy import *
import sys


a = 0.7  
b = 0.5 
c = 0.3  
e = 0.2
```

# Parámetros del simulación:

- **Paso de tiempo para la simulación (dt):** 0.001
- **Tiempo máximo para ejecutar la simulación (max_time):** 100





```python
dt = 0.001      
max_time = 100  
```

# Condiciones iniciales

- **Tiempo inicial (t):** 0
- **Población inicial de presas (x):** 1.0
- **Población inicial de depredadores (y):** 0.5

```python
t = 0        
x = 1.0      
y = 0.5      
```

# Listas vacías para almacenar el tiempo y las poblaciones para su futura graficación
- **Lista para almacenar los tiempos** (t_list)
- **Lista para almacenar la población de presas en cada instante** (x_list)
- **Lista para almacenar la población de depredadores en cada instante** (y_list)

### Se almacenan valores iniciales

```python
t_list = []  
x_list = []  
y_list = []  


t_list.append(t)  
x_list.append(x) 
y_list.append(y)  
```

# Bucle para ejecutar la simulación hasta que el tiempo alcance max_time, actualizando los valores en cada iteración

1. **t = t + dt**: Avanzar el tiempo en `dt` unidades.
2. **x = x + (a*x - b*x*y) * dt**: Actualizar la población de presas usando la ecuación de Lotka-Volterra.
3. **y = y + (-c*y + e*x*y) * dt**: Actualizar la población de depredadores usando la ecuación de Lotka-Volterra.
4. **Se almacenan los datos en las listas creadas con anterioridad.**
5. **Se usa `matplotlib` por medio de `plt` para graficar:**
   - **r =** Graficar la población de presas en color rojo.
   - **g =** Graficar la población de depredadores en color verde.
   - **Grueso de la línea de la gráfica** (`linewidth`)=2.
   - **`plt.xlabel('Tiempo')`**: Etiqueta para el eje x.
   - **`plt.ylabel('Población')`**: Etiqueta para el eje y.
   - **`plt.title('Modelo de Lotka-Volterra: Presas vs Depredadores')`**: Título del gráfico.
   - **`plt.show()`**: Mostrar el gráfico.


```python
while t < max_time:
    t = t + dt
    x = x + (a*x - b*x*y) * dt
    y = y + (-c*y + e*x*y) * dt

    t_list.append(t)
    x_list.append(x)
    y_list.append(y)

p = plt.plot(t_list, x_list, 'r',
             t_list, y_list, 'green',
             linewidth=2)
plt.xlabel('Tiempo')
plt.ylabel('Población')
plt.title('Modelo de Lotka-Volterra: Presas vs Depredadores')
plt.show()
```

```python
while t < max_time and x > 0 and y > 0:
    t = t + dt
    x = x + (a*x - b*x*y) * dt
    y = y + (-c*y + e*x*y) * dt

    t_list.append(t)
    x_list.append(x)
    y_list.append(y)

p = plt.plot(t_list, x_list, 'r',
             t_list, y_list, 'green',
             linewidth=2)
plt.xlabel('Tiempo')
plt.ylabel('Población')
plt.title('Modelo 5 de Lotka-Volterra: Presas vs Depredadores')
plt.show()
 
```
