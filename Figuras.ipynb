# Se importan las librerías

```python
import matplotlib.pyplot as plt
```

# Se definen los intervalos del plano de 0 a 1 tanto en x como en y.

```python

plt.plot([1],[1])
```

### Explicación del Código:
- **`plt`**: Se refiere a la interfaz de Pyplot de Matplotlib, que se utiliza para crear gráficos en 2D de forma sencilla y eficiente.

- **`plot()`**: Esta función es utilizada para dibujar una línea en el gráfico. Recibe dos listas como argumentos:

  - **Primer argumento: `[1, 1.5, 2]`**: Esta lista representa las coordenadas en el eje x de los puntos que se desean graficar.
  
  - **Segundo argumento: `[1, 2, 1]`**: Esta lista representa las coordenadas en el eje y de los puntos correspondientes a las coordenadas en el eje x.


```python
plt.plot([1,1.5,2],[1,2,1])
```

### Adición de Coordenadas en el Gráfico

Se añaden dos coordenadas más que indican el recorrido desde el último punto **(2, 1)** a **(1, 1)**. Esta conexión adicional forma una línea que vuelve al punto (1, 1), cerrando así el recorrido en el gráfico.


```python
plt.plot([1,1.5,2,1],[1,2,1,1])
```

### Generación de un Cuadrado

Para generar un cuadrado en un gráfico se define el siguiente recorrido

(1,1) a (1,2) a (2,2) a (2,1) a (1,1)




```python
plt.plot([1,1,2,2,1],[1,2,2,1,1])
```

### Generación de un Cuadrado con un Triángulo Interno

- **Cuadrado**:
  - **`plt.plot([1, 1, 2, 2, 1], [1, 2, 2, 1, 1])`**: Este comando genera un cuadrado con los puntos indicados.
  
- **Triángulo**:
  - **`plt.plot([1, 1.5, 2], [1, 2, 1])`**: Este comando se utiliza para graficar un triángulo dentro del cuadrado.
    - **Coordenadas X**: `[1, 1.5, 2]`
      - (1, 1): Punto inferior izquierdo del triángulo
      - (1.5, 2): Punto superior del triángulo
      - (2, 1): Punto inferior derecho del triángulo

```python
plt.plot([1,1,2,2,1],[1,2,2,1,1])
plt.plot([1,1.5,2],[1,2,1])
```
